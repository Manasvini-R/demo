```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from lightgbm import LGBMClassifier

# Load dataset
df = pd.read_csv("malicious_phish.csv")

# Reduce dataset size for faster training (stratified sampling)
df_sampled = df.groupby("type").apply(lambda x: x.sample(min(len(x), 5000)))

# Extract features and labels
X = df_sampled["url"]
y = df_sampled["type"]

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(ngram_range=(1, 3), max_features=5000)
X_vectorized = vectorizer.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y_encoded, test_size=0.2, random_state=42)

# Train LightGBM model (Faster and More Accurate)
model = LGBMClassifier(n_estimators=200, max_depth=10, learning_rate=0.1, random_state=42)
model.fit(X_train, y_train)

# Save model, vectorizer, and label encoder
joblib.dump(model, "model.pkl")
joblib.dump(vectorizer, "vectorizer.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")

print("Model, vectorizer, and label encoder saved successfully!")
```
